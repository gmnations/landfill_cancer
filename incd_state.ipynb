```python
# importing Dependencies
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Reading/importing csv file
# Printing csv file
incd_csv = "resources/incd_state.csv"
incd_df = pd.read_csv(incd_csv, keep_default_na=False)
incd_df.head()
```

```python
# Removing the first to the seventh rows using the drop function
# Printing the changed DataFrame
incd1_df = incd_df.drop(incd_df.index[0:7])
incd1_df.head()
```

```python
# Resetting the index
# Printing the changed DataFrame
incd2_df = incd1_df.reset_index(drop=True)
incd2_df.head()
```

```python
#Accessing the row of column headers with pandas.DataFrame.iloc
#Resetting the row of column headers using the values found in the first row
#Resetting the index again after previous changes
#Printing the changed DataFrame
new_header = incd2_df.iloc[0] 
incd2_df = incd2_df[1:] 
incd2_df.columns = new_header
incd3_df = incd2_df.reset_index(drop=True)
incd3_df.head()
```

```python
#Removing the first row using the drop function
#Printing the changed DataFrame
incd4_df = incd3_df.drop(incd3_df.index[0])
incd4_df.head()
```

```python
# Resetting the index after the changes
#Printing the changed DataFrame
incd5_df = incd4_df.reset_index(drop=True)
incd5_df.head()
```

```python
# Dropping unwanted rows at the bottom of the DataFrame
incd6_df = incd5_df.drop(incd5_df.index[52:86])
```

```python
# Removing the numbers and parentheses after the strings in the "State" column
# Renaming the "Age-Adjusted Incidence Rate([rate note]) - cases per 100,000" to "Cases per 100,000"
# Removing the # sign that was on one of the Cases per 100,000 line
# Printing the changed DataFrame
incd6_df['State'] = incd6_df['State'].str.replace("(","").replace(")","").str.replace("5","").str.replace("6","").str.replace("7","").str.replace(")","")
incd6_df.rename(columns = {'Age-Adjusted Incidence Rate([rate note]) - cases per 100,000':'Cases per 100,000'}, inplace = True)
incd6_df['Cases per 100,000'] = incd6_df['Cases per 100,000'].str.replace("#", "")
incd6_df.head()
```

```python
# Replacing the states values found in the State column with the abbreviations from the dictionary
# Priting the changed DataFrame
incd6_df = incd6_df.replace({'State': us_state_abbrev})
```

```python
incd7_df = incd6_df[["State", "Cases per 100,000"]]
incd7_df.sort_values("Cases per 100,000",inplace=True, ascending=False)
incd7_df.head()
```

```python
# Printing the new edited csv file
incd7_df.to_csv("cancer_cases_per_state.csv")
```

```python
# Filtering the DataFrame down only to those columns to chart
# Setting the index to be "State" so they will be used as labels
state_and_cancer = incd7_df
state_and_cancer = state_and_cancer.set_index("State")
state_and_cancer.head()
```

```python
# Converting my data into floats because it will not recognize the numeric values when plotting otherwise 
# Using DataFrame.plot() in order to create a bar chart of the data
# Setting a title for the chart
# Setting labels
state_and_cancer = state_and_cancer.astype(float)
state_and_cancer.plot(kind="bar", figsize = (20,3))
plt.title("Cancer cases per 100K by state")
plt.xlabel("State")
plt.ylabel("Cases per 100,000")
plt.tight_layout()
plt.savefig("images/cancer_per_state.png")
```

```python
# Reading/importing the State Population csv file
# Printing  the State Population csv file
population_csv = "resources/state_populations_2000_to_2019.csv"
population_df = pd.read_csv(population_csv, keep_default_na=False)
population_df.head()
```

```python
# Replacing the state_name values with the abbreviations from the dictionary
# Printing the changed DataFrame
population_df = population_df.replace({'state_name': us_state_abbrev})
population_df.head()
```

```python
# Listing all columns that are in the DataFrame
population_df.columns
```

```python
# Keeping the needed columns
# Printing the changed DataFrame with the remaining columns
cleaned_population = population_df[['state_name', 'pop_2013', 'pop_2014', 'pop_2015', 'pop_2016', 'pop_2017']]
cleaned_population.head()
                                
```

```python
# Calculating the Average Population and inputting that into a new column
# Printing the changed DataFrame with the new column
cleaned_population['Average_Population'] = (cleaned_population.pop_2013 + cleaned_population.pop_2014 + cleaned_population.pop_2015 + cleaned_population.pop_2016 + cleaned_population.pop_2017) / 5
cleaned_population.head()
```

```python
# Removing the columns that are not needed by keeping only the ones that are needed after calculating the Average
# Printing the changed DataFrame
cleaned_population = cleaned_population[['state_name', 'Average_Population']]
cleaned_population.head()
```

```python
# Renaming the state_name column to State
# Printing the changed DataFrame
cleaned_population.rename(columns = {'state_name':'State'}, inplace=True)
cleaned_population.head()
```

```python
# Sorting the State Values alphabetically
cleaned_population.sort_values("State", inplace=True, ascending=True)
```

```python
# Calculating the Average Population per 100K and adding this data into a new column
# Printing the changed DataFrame
cleaned_population['Population_per_100K'] = (cleaned_population.Average_Population) / 100000
cleaned_population.head()
```

```python
# Merging the cancer data and the Population data on the common state column
# Printing the merged DataFrame
cancer_and_population = pd.merge(cleaned_population, incd7_df, on="State")
cancer_and_population.head()
```

```python
# Renaming the Cases per 100,000 column
# Printing the changed DataFrame
cancer_and_population.rename(columns = {'Cases per 100,000':'Cases_per_100K'}, inplace=True)
cancer_and_population.head()
```

```python
# Changing the datatype into floats in order to calculate the Total Cases
# Calculating the Total Cases
# Printing the changed DataFrame
cancer_and_population[['Population_per_100K', 'Cases_per_100K']] = cancer_and_population[['Population_per_100K', 'Cases_per_100K']].astype(float)
cancer_and_population['Total_Cases'] = cancer_and_population['Population_per_100K'] * cancer_and_population['Cases_per_100K']
cancer_and_population.head()
```

```python
# Sorting the Total Cases values
# Printing the changed DataFrame
sorted_df = cancer_and_population.sort_values("Total_Cases", ascending=False)
sorted_df.head()
```

```python
# Plotting the data by State and by Total Cases and Cases per 100K
# Setting a title for the chart
# Setting labels
sorted_df.plot(x="State", y=["Total_Cases","Cases_per_100K"], kind="bar")
plt.rcParams['figure.figsize'] = (25,5)
plt.title('Total Cases and Cases Per 100K by State')
plt.xlabel('State')
plt.ylabel('Cases')
plt.show()
```

```python
# Saving the plot image
plt.savefig("images/cancer_and_population.png")
```

```python
# Resetting the index of the DataFrame
# Exporting the new DataFrame into a new csv file
sorted_df = sorted_df.reset_index(drop=True)
sorted_df.to_csv("cancer_and_population.csv")
```
