```python
import pandas as pd
import numpy as np
```

```python
read_file = "Resources/incd.csv"
incd_df = pd.read_csv(read_file, keep_default_na=False, skiprows = 8)
incd_df.dropna(inplace = True)
incd_df.head(20)
```

```python
new_header = incd_df.iloc[0] 
incd_df = incd_df[1:] 
incd_df.columns = new_header
incd3_df = incd_df.reset_index(drop=True)
incd3_df.head()
```

```python
incd4_df = incd3_df.drop(incd3_df.index[0])
incd4_df.head()
```

```python
incd5_df = incd4_df.reset_index(drop=True)
incd5_df.head()
```

```python
incd6_df = incd5_df.drop(incd5_df.index[3141:])



incd6_df.tail()
```

```python
incd6_df['State'] = incd6_df['State'].str.replace("(","").replace(")","").str.replace("5","").str.replace("6","").str.replace("7","").str.replace(")","")
incd6_df.head()
```

```python
new = incd6_df["State"].str.split(",", n = 1, expand = True)
```

```python
new
```

```python
new["County"]= new[0]
new["State"]= new[1]
```

```python
incd7_df = pd.merge(new,incd6_df, left_index=True, right_index=True)
```

```python
incd7_df
```

```python
incd7_df.drop(columns=[0, 1], inplace=True)
```

```python
incd8_df = incd7_df.drop(columns=['Upper 95% Confidence Interval','Lower 95% Confidence Interval','Recent 5-Year Trend ([trend note]) in Incidence Rates','Upper CI (CI*Rank)','Lower CI (CI*Rank)','CI*Rank([rank note])','Upper 95% Confidence Interval','Lower 95% Confidence Interval','Met Healthy People Objective of ***?'])
```

```python
incd8_df
```

```python
incd8_df.to_csv("eEdited_incd.csv")
```
